**DATA CLEANING**

```python
charge_cols_raw = [
    'Average Covered Charges',
    'Average Total Payments',
    'Average Medicare Payments'
]

for col in charge_cols_raw:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )
```

```python
df[charge_cols_raw] = df[charge_cols_raw].apply(
    pd.to_numeric,
    errors='coerce'
)
```

```python
df[charge_cols_raw].head()
```

```python
df_clean = df.copy()
```

```python
df_clean.columns = (df_clean.columns
                    .str.lower()
                    .str.strip()
                    .str.replace(' ', '_'))

df_clean.columns
```

```python
df_clean = df_clean.rename(columns={
    "drg_definition": "drg",
    "provider_id": "provider_id",
    "provider_name": "hospital_name",
    "provider_street_address": "address",
    "provider_city": "city",
    "provider_state": "state",
    "provider_zip_code": "postal_code",
    "hospital_referral_region_(hrr)_description": "hospital_referral_region",
    "total_discharges": "total_discharges",
    "average_covered_charges": "avg_covered_charges",
    "average_total_payments": "avg_total_payments",
    "average_medicare_payments": "avg_medicare_payments"
})
```

```python
df_clean.dtypes
```

```python
charge_cols = [
    "avg_covered_charges",
    "avg_total_payments",
    "avg_medicare_payments"
]

for col in charge_cols:
  df_clean[col] = (
      df_clean[col]
      .astype(str)
      .str.replace('$', '', regex=True)
      .str.replace(',', '', regex=True)
      .str.strip()
      )
```

```python
df_clean[charge_cols] = df_clean[charge_cols].apply(
    pd.to_numeric,
    errors='coerce'
)
```

```python
df_clean[charge_cols].dtypes
```

```python
df_clean["total_discharges"].value_counts().head(10)
```

```python
df_clean["total_discharges"] = (
    df_clean["total_discharges"]
    .astype(str)
    .str.strip()
)
```

```python
df_clean["total_discharges"] = pd.to_numeric(
    df_clean["total_discharges"],
    errors="coerce"
)
```

```python
df_clean = df_clean[df_clean["total_discharges"] > 0]
df_clean['total_discharges'] = df_clean['total_discharges'].astype(int)
```

```python
df_clean.dtypes
```

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean["drg_code"] = df_clean["drg"].str.split(' - ').str[0]
```

```python
df_clean["drg_description"] = df_clean["drg"].str.split(' - ').str[1]
```

```python
df_clean[charge_cols].head()
df_clean[charge_cols].isna().sum()
df_clean.dtypes
```

```python
df_clean['provider_id'] = df_clean['provider_id'].astype(int)
```

```python
df_clean['postal_code'] = df_clean['postal_code'].astype(int).astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

**Distribution Of Average Covered Charges**

```python
plt.figure()
sns.histplot(df_clean['avg_covered_charges'], bins=50, kde=True)
plt.title('Distribution Of Average Covered Charges')
plt.xlabel('Average Covered Charges')
plt.ylabel('Frequency')
plt.show()
```

**INTERPRETATION**

* Strong Right-Skew
* Most DRGs have moderate charges
* Few DRGs have extremely high charges



**Charges Vs Medicare Payments**

```python
plt.figure()
sns.boxplot(data=df_clean[['avg_covered_charges', 'avg_medicare_payments']])
plt.title('Hospital Charges Vs Medicare Payments')
plt.ylabel('Amount')
plt.show()
```

**INTERPRETATION**

* Medicare payments are far lower than charges
* Both have outliers, but charges vary much more
* Confirms reimbursement gap

**State-wise Average Covered Charges (Top 10)**

```python
state_avg = (
    df_clean.groupby('state')['avg_covered_charges']
            .mean()
            .sort_values(ascending=False)
            .head(10)
)
```

```python
plt.figure()
state_avg.plot(kind='bar')
plt.title('Top 10 States By Average Covered Charges')
plt.xlabel('State')
plt.ylabel('Average Covered Charges')
plt.show()
```

**INTERPRETATION**

* Certain states consistently show higher charges
* Influenced by:

      * Cost of living
      * Teaching hospitals
      * Procedure complexity

**Total Charges Vs Average Covered Charges**

```python
plt.figure()
sns.scatterplot(
    x='total_discharges',
    y='avg_covered_charges',
    data=df_clean,
    alpha=0.4
)

plt.title('Discharges Vs Average Covered Charges')
plt.xlabel('Total Discharges')
plt.ylabel('Average Covered Charges')
plt.show()
```

**INTERPRETATION**

* No strong linear relationship
* High-cost procedures are often low volume
* High-volume DRGs tend to be lower cost

**Medicare Payment Ratio**

```python
df_clean['medicare_payment_ratio'] = (
    df_clean['avg_medicare_payments'] / df_clean['avg_covered_charges'] * 100
)
```

```python
plt.figure()
sns.histplot(df_clean['medicare_payment_ratio'], bins=40, kde=True)
plt.title('Medicare Payment as % of Hospital Charges')
plt.xlabel('Medicare Payment Ratio')
plt.ylabel('Frequency')
plt.show()
```

**INTERPRETATION**

* Most ratios are well below 1
* Medicare typically pays 20-40% of charges
* Shows reimbursement pressure clearly
